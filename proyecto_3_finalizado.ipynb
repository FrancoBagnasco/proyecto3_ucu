## <u>Tasa de Desempleo (%) por Rango Etario en Uruguay (2006-2017)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Cargar el dataset desde un archivo XLSX
df = pd.read_excel('data/tasadesempleo.xlsx')

# Crear un gráfico de líneas para cada rango de edad
plt.figure(figsize=(12, 6))

for rango_edad in df['Tramo de Edad'].unique():
    data_filtrada = df[df['Tramo de Edad'] == rango_edad]
    plt.plot(data_filtrada['año'], data_filtrada['valor'], label=rango_edad)

plt.xlabel('Año')
plt.ylabel('Tasa de Desempleo (%)')
plt.title('Tasa de Desempleo (%) por Rango Etario en Uruguay (2006-2017)')
plt.legend()
plt.grid(True)

# Mostra los años de uno en uno en el eje X
plt.xticks(df['año'].unique(), rotation=45)

# Mostrar el gráfico
plt.show()
```

### Pregunta de investigación

La visualización de la tasa de desempleo por rango etario en Uruguay (2006-2017) revela tendencias interesantes en el mercado laboral. Para profundizar en este análisis, podemos plantear una pregunta de investigación:

**¿Cómo ha variado la tasa de desempleo en diferentes grupos de edad en Uruguay durante el período de 2006 a 2017 y cuáles son las tendencias más notables?**

### Análisis de la gráfica

La gráfica muestra la variación de la tasa de desempleo para cinco rangos de edad diferentes: "14 a 24", "25 a 29", "30 a 49", "50 a 64" y "65 y más", durante 11 periodos de 1 año. Aquí hay algunas observaciones iniciales:

* La tasa de desempleo es más alta en el rango de edad "14 a 24" durante todo el período.
* En general, se observa una tendencia a la disminución de la tasa de desempleo a medida que aumenta el rango de edad.
* El rango de edad "65 y más" tiende a tener una tasa de desempleo relativamente baja y estable.
* Todas los rangos etarios experimentaron su **nivel máximo de desempleo en el 2006.**
* Todos los rangos etarios experimentaron sus **niveles mínimos de desempleo entre 2011 y 2012**, luego comenzaron a aumentar nuevamente.

## <u>Tasa de Desempleo (%) del Rango Etario 15-24 en Panamá, Chile y Costa Rica (2006-2017)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Carga el archivo Excel
df = pd.read_excel('data/desempleopanama1524.xls')

# Filtra los datos de Panamá, Chile y Costa Rica y selecciona las columnas de 2006 a 2017
countries_data = df[(df['Country Name'].isin(['Panama', 'Chile', 'Costa Rica'])) & (df['Country Code'].isin(['PAN', 'CHL', 'CRI']))]
countries_data = countries_data[['Country Name', 'Indicator Name', 'Indicator Code', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']]

# Transponer los datos para que los años estén en una columna
countries_data = countries_data.melt(id_vars=['Country Name', 'Indicator Name', 'Indicator Code'], var_name='Año', value_name='Tasa de Desempleo')

# Convierte la columna de 'Año' a tipo numérico
countries_data['Año'] = countries_data['Año'].astype(int)

# Crea la gráfica
plt.figure(figsize=(10, 6))
for country in ['Panama', 'Chile', 'Costa Rica']:
    data_country = countries_data[countries_data['Country Name'] == country]
    plt.plot(data_country['Año'], data_country['Tasa de Desempleo'], marker='o', label=country)

plt.xlabel('Año')
plt.ylabel('Tasa de Desempleo (%)')
plt.title('Tasa de Desempleo (%) del Rango Etario 15-24 en Panamá, Chile y Costa Rica (2006-2017)')
plt.grid(True)
plt.legend()

# Mostrar los años de uno en uno en el eje X
plt.xticks(countries_data['Año'], rotation=45)

# Mostrar el gráfico
plt.show()
```

### Pregunta de investigación

La visualización de la tasa de desempleo de los más jovenes en estos 3 paises (en el periodo 2006-2017) puede ayudarnos a determinar que tan bueno o malo es el comportamiento de dicha métrica en Uruguay. Al compararlo con paises similares de la región podemos empezar a sacar muchas más conclusiones válidas que las que podriamos obtener simplemente mirando la primera gráfica. ### Análisis de la gráfica

La gráfica muestra la variación de la tasa de desempleo (del rango etario 15-24) para 3 paises de la region con aspectos macro similares a los de Uruguay, por ejemplo:
* **Estos paises tienen los PBI per cápita más altos de latinoamerica.**

## <u>Tasa de Desempleo (%) del Rango Etario 14-24 en Uruguay, Panamá, Chile y Costa Rica (2006-2017)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Cargar el dataset de Uruguay
df_uruguay = pd.read_excel('data/tasadesempleo.xlsx')

# Filtrar los datos para mostrar solo el rango de edad "14 a 24"
df_uruguay_14_24 = df_uruguay[df_uruguay['Tramo de Edad'] == '14 a 24']

# Crear un gráfico de líneas para el rango de edad "14 a 24" en Uruguay
plt.figure(figsize=(12, 6))
plt.plot(df_uruguay_14_24['año'], df_uruguay_14_24['valor'], marker='o', label='Uruguay')

# Cargar el dataset de los comparativos
df_comparativos = pd.read_excel('data/desempleopanama1524.xls')

# Filtrar las columnas de interés (2006 a 2017)
df_comparativos = df_comparativos[['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']]

# Transponer los datos para que los años estén en una columna
df_comparativos = df_comparativos.melt(id_vars=['Country Name'], var_name='Año', value_name='Tasa de Desempleo')

# Convierte la columna de 'Año' a tipo numérico
df_comparativos['Año'] = df_comparativos['Año'].astype(int)

# Agregar los datos de Panamá, Chile y Costa Rica al gráfico
for country in ['Panama', 'Chile', 'Costa Rica']:
    data_country = df_comparativos[df_comparativos['Country Name'] == country]
    plt.plot(data_country['Año'], data_country['Tasa de Desempleo'], marker='o', label=country)

plt.xlabel('Año')
plt.ylabel('Tasa de Desempleo (%)')
plt.title('Tasa de Desempleo (%) para el Rango Etario 14-24 en Uruguay y Países Comparativos (2006-2017)')
plt.legend()
plt.grid(True)

# Mostrar los años de uno en uno en el eje X
plt.xticks(countries_data['Año'], rotation=45)

# Mostrar el gráfico
plt.show()
```

### Análisis de la gráfica

La gráfica muestra la variación de la tasa de desempleo (del rango etario 14-24) para los 4 paises de la region analizados en el proyecto, esto nos permite resaltar ciertas observaciones interesantes:
* **Panamá ha sido con diferencia el país más estable** y con menor desempleo juvenil durante casi todo el periodo.
* A Panamá le sigue Costa Rica en estabilidad y menor desempleo en la última mitad del periodo.
* **Uruguay comienza y termina el periodo como el país con mayor desempleo**, sin embargo, no mostro variaciones tan bruzcas año a año como Chile.
* También se observa claramente que **2 paises sufrieron** (en mayor o menor medida) **la crisis del 2008** ya que tuvieron grandes aumentos en el desempleo en el periodo 2008-2009, estos fueron Chile y Costa Rica.
* En el 2011, **Uruguay logró posicionarse brevemente como el 2 país con menor desempleo juvenil** de los 4, solo superado por Panamá.
* Todos los paises experimentaron una subida en el desempleo en los últimos 4 años a excepcion de Chile.
* Chile experimentó la mayor subida del desempleo en un año, aprox. **+ 7%** en el periodo 2008-2009.
* Costa Rica experimetó la mayor bajada del desempleo en un año, aprox. **- 7%** en el periodo 2009-2010.

### Conclusión

El análisis detallado de la tasa de desempleo juvenil en Uruguay y otros tres países de la región para el período 2006-2017 arroja luz sobre las tendencias y dinámicas en el mercado laboral que afectan a los jóvenes. A través de esta exploración, se pueden extraer conclusiones significativas que proporcionan una comprensión más profunda de la situación laboral de esta población en los países estudiados.

Este estudio ademas, nos muestra que el mercado laboral uruguayo es más inflexible que el del resto de paises comparados, lo que puede derivar en tasas mayores de desempleos e incluso a un número menor de emprendimientos al largo plazo debido a las dificultades de la contratación y el problema de limitar el crecimiento al no poder ampliar la plantilla facilmente.

La contracara de este hecho, es que desde el punto de vista del empleado, es probable que se logre un salario promedio mayor en Uruguay, lo que favorece a los contratados pero disminuye las probabilidades de ser contratado para los desempleados. Es decir, desde el punto de vista de quienes buscan empleo, la situación puede ser más desafiante, ya que la rigidez del mercado laboral puede limitar las oportunidades laborales y hacer que la competencia por los puestos de trabajo disponibles sea más feroz. Esto puede ser particularmente relevante para los jóvenes que ingresan al mercado laboral, ya que enfrentan tasas de desempleo más altas en comparación con otros grupos de edad.

Otro punto importante a destacar es la variabilidad que se observa en las tasas de desempleo a lo largo del período. En el caso de Chile, se destacó por tener la variación más notable en las tasas de desempleo juvenil. El país experimentó un aumento significativo de aproximadamente el 7% en el período 2008-2009, seguido por disminuciones y aumentos en los años subsiguientes. Esta variabilidad podría atribuirse a factores económicos y políticos específicos que influyeron en el mercado laboral chileno. Costa Rica solo pudo estabilizarse en los últimos años, mientras que Uruguay fue más estable constantemente durante todo el periodo, lo que se puede confirmar al saber que Uruguay no sufrio crisis en dicho periodo.

Por otro lado, Costa Rica experimentó la mayor disminución del desempleo en un año, con una reducción de aproximadamente el 7% en el período 2009-2010. Esto puede relacionarse con políticas y medidas específicas tomadas por el gobierno costarricense para abordar la situación del mercado laboral.

En el caso de Uruguay, a pesar de comenzar y finalizar el período con las tasas de desempleo más altas, el país mostró una variación más gradual a lo largo de los años. No obstante, en 2011, Uruguay logró posicionarse brevemente como el segundo país con la tasa de desempleo juvenil más baja entre los cuatro países analizados, solo superado por Panamá.

En resumen, el análisis detallado de la tasa de desempleo juvenil en Uruguay y los países comparativos revela una serie de dinámicas y tendencias interesantes en el mercado laboral. A través de la observación de la inflexibilidad del mercado laboral uruguayo, las respuestas a la crisis económica global y las variaciones en las tasas de desempleo a lo largo del período, es posible obtener una comprensión más profunda de los factores que afectan a los jóvenes en busca de empleo en la región.
